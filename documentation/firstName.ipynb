# Test de chi2 
--------------
On suppose qu'on dispose de n variables aléatoires $X_i$ discrètes, à valeurs dans ${a_1,…,a_k}$, indépendantes et de même loi caractérisée par p=($p_1$,…,$p_k$) où $p_i=P(X=ai)$

On souhaite tester :
  $H_0=p_0$
  $H_1≠p_0$

où $p_0=(p_{01},…,p_{0k})$ est connue.
On utilise à cet effet le test du Khi-deux avec comme statistique :

$\sum n=n \sum_{i=1}^k\frac{(p^{i}−p_{0}^{i})^{2}}{p_{0}^{i}}$

où $p^{i}$ est égal à la fréquence empirique de la valeur $a_{i}$.

La loi asymptotique de $\sum n$
sous $H_0$ est connue, il s'agit d'une loi du Khi-deux à $k−1$ degrés de liberté. 

```python
all_gram = test.compute_all_gram("DicoFirstName.ref")   
file_referential  =["DicoLastName.ref" ,"DicoFirstName.ref","DicoLocality.ref"]
features , target = test.get_feature_chi2(file_referential,all_gram)
X = np.array(features)
Y = np.array(target)
liberty_degree =  (X[0].size-1)
critic_value = scipy.stats.chi2.ppf(q = 0.95,df = liberty_degree) 
stat = test.compute_khi2(X,Y,1)
pl.plot(stat[0])
```
